```python
#importing libraries
from pandasql import sqldf #pandasql is a library that allows you to write SQL queries directly on pandas DataFrames.
import pandas as pd
```

Lets create a lambda function pysqldf acts as a shortcut or alias for executing SQL queries on pandas DataFrames using the sqldf function

```python
pysqldf = lambda q: sqldf(q, globals())
```

```python
#importing the csv file 
df1=pd.read_csv('dataset1.csv')
df1.head(10)
```

# Table 1: QUERIES USING PANDAS SQL

### 1) Display all the details of movies created by directors Christopher and Matt Reeves.

```python
Christopher_result="select * from df1 where `Director 1` like ('%Christopher Nolan%')"
chris=pysqldf(Christopher_result)
print(len(chris))
chris.head(10)
```

```python
reeves_result="select * from df1 where `Director 1` like ('%Matt Reeves%')"
details=pysqldf(reeves_result)
print(len(details))
details.head(10)
```

### 2) Display all the details of movies with a duration of 140 minutes to 190 minutes.

```python
du="select * from df1 where Duration >=140 and Duration <= 190"
duration=pysqldf(du)
print(len(duration))
duration.head(10)
```

### 3) Display all details of movies with ratings above 7 in ascending order.

```python
rat="select * from df1 where `Ratings` >7 order by Ratings asc"
ratings=pysqldf(rat)
print(len(ratings))
ratings.head(10)
```

```python
mv="select `Movie Name` from df1  order by `Movie Name` desc"
descending=pysqldf(mv)
print(len(descending))
descending.head(10)
```
